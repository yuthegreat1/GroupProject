```python
#dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading in csv files
netflix_csv = "../CSV Files/netflix_titles.csv"
hulu_csv = "../CSV Files/hulu_titles.csv"
amazonprime_csv = "../CSV Files/amazon_prime_titles.csv"
disneyplus_csv = '../CSV Files/disney_plus_titles.csv'
```

```python
#reading files into dataframes
netflix_df = pd.read_csv(netflix_csv)
hulu_df = pd.read_csv(hulu_csv)
amazonprime_df = pd.read_csv(amazonprime_csv)
disneyplus_df = pd.read_csv(disneyplus_csv)
```

```python
#showing previews of each data set
netflix_df.head()
```

```python
hulu_df.head()
```

```python
amazonprime_df.head()
```

```python
disneyplus_df.head()
```

```python
#grouping each streaming platform by type of media, either movie or tv show
grouped_netflix = netflix_df.groupby(["type"])
grouped_hulu = hulu_df.groupby(["type"])
grouped_amazon = amazonprime_df.groupby(["type"])
grouped_disney = disneyplus_df.groupby(["type"])
grouped_netflix.count()
```

```python
#saving x & y axis for a plot
yaxis_net = grouped_netflix["title"].count()
xaxis_net = np.arange(len(yaxis_net))
plt.bar(xaxis_net, yaxis_net, color = 'r', align="center")
tick_locations = [value for value in xaxis_net]
plt.xticks(tick_locations, ["Movie", "TV Show"])
plt.title("Netflix Content by Type")
plt.xlabel("Type of Content")
plt.ylabel("Count")
```

```python
#creating plot for hulu
yaxis_hulu = grouped_hulu["title"].count()
xaxis_hulu = np.arange(len(yaxis_hulu))
plt.bar(xaxis_hulu, yaxis_hulu, color = 'g', align="center")
tick_locations = [value for value in xaxis_hulu]
plt.xticks(tick_locations, ["Movie", "TV Show"])
plt.title("Hulu Content by Type")
plt.xlabel("Type of Content")
plt.ylabel("Count")
```

```python
#creating plot for amazon
yaxis_amazon = grouped_amazon["title"].count()
xaxis_amazon = np.arange(len(yaxis_amazon))
plt.bar(xaxis_amazon, yaxis_amazon, color = 'y', align="center")
tick_locations = [value for value in xaxis_amazon]
plt.xticks(tick_locations, ["Movie", "TV Show"])
plt.title("Amazon Video Content by Type")
plt.xlabel("Type of Content")
plt.ylabel("Count")
```

```python
#creating a plot for Disney+
yaxis_disney = grouped_disney["title"].count()
xaxis_disney = np.arange(len(yaxis_disney))
plt.bar(xaxis_amazon, yaxis_amazon, color = 'b', align="center")
tick_locations = [value for value in xaxis_disney]
plt.xticks(tick_locations, ["Movie", "TV Show"])
plt.title("Disney+ Content by Type")
plt.xlabel("Type of Content")
plt.ylabel("Count")
```

```python
netflix_df['date_added'].head()
```

```python
#Splitting 'date_added' column into two columns
netflix_df[['Date Added Month', 'Date Added Year']] = netflix_df['date_added'].str.split(', ', expand=True)
netflix_df.head()
```

```python
#for hulu
hulu_df[['Date Added Month', 'Date Added Year']] = hulu_df['date_added'].str.split(', ', expand=True)
hulu_df.head()
```

```python
#for amazon
amazonprime_df[['Date Added Month', 'Date Added Year']] = amazonprime_df['date_added'].str.split(', ', expand=True)
amazonprime_df.head()
```

```python
#for disney+
disneyplus_df[['Date Added Month', 'Date Added Year']] = disneyplus_df['date_added'].str.split(', ', expand=True)
disneyplus_df.head()
```

```python
#grouping by the type of content and the yeard of upload
netflix2021 = netflix_df.groupby('type')['Date Added Year'].value_counts().sort_index()
hulu2021 = hulu_df.groupby('type')['Date Added Year'].value_counts().sort_index()
amazon2021 = amazonprime_df.groupby('type')['Date Added Year'].value_counts().sort_index()
disney2021 = disneyplus_df.groupby('type')['Date Added Year'].value_counts().sort_index()
```

```python
#showing a preview to see which years are relevant
netflix2021
```

```python
#saving years as variable
years_net = range(2008,2022)
```

```python
#saving by content
net_movie_up = netflix2021['Movie']
net_tv_up = netflix2021['TV Show']
```

```python
#adding 0s to empty years
net_tv_up['2009']=0
net_tv_up['2010']=0
net_tv_up['2011']=0
net_tv_up['2012']=0
net_tv_up = net_tv_up.sort_index(ascending=True)
net_tv_up
```

```python
#creating a line graph to show movie vs tv uploads by year for Netflix
plt.plot(years_net, net_movie_up, color="red", label="Movies" )
plt.plot(years_net, net_tv_up, color='black', label="TV Shows")
plt.legend()
plt.title("Netflix Uploads by Type & Year")
plt.show()
```
