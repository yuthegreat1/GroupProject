```python
#subset netflix dataset
netflix_origs = netflix_origs[["Title", "IMDB Score", "Language"]]
```

```python
#count movies 1 time for each language they are available in
lst = []
for i in range(len(netflix_origs)):
    language = str(netflix_origs.loc[i]["Language"])
    if ("/" in language):
        split = language.split("/")
        for j in range(len(split)):
            row = netflix_origs.loc[i].copy()
            row["Language"] = split[j]
            lst.append(row)
    else:
        row = netflix_origs.loc[i].copy()
        lst.append(row)
netflix_origs_split = pd.DataFrame(lst)
netflix_origs_split
```

```python
language = netflix_origs_split.groupby("Language")
languagemean = pd.DataFrame(language.mean()).sort_values(["IMDB Score"], ascending = False)
#sort by most popular languages used in netflix 
languagesorted = netflix_origs_split.sort_values(["IMDB Score"], ascending = False)
langsortedhead = languagesorted.head()
language.count()
```

```python
plt.scatter(languagesorted["Language"], languagesorted["IMDB Score"], color='lightblue', edgecolor='blue')
plt.xticks(rotation = 90)
plt.xlabel("Language")
plt.ylabel("IMDB Rating")
plt.title("IMDB Ratings of Netflix Original Movies According to Language")
#save fig
plt.savefig("Figures/language_popularity.png")
```

```python
#find means of movie ratings according to language
language.mean()
languagemean = pd.DataFrame(language.mean()).sort_values(["IMDB Score"], ascending = False)
languagemean
#save languagemean as the means of the top 5 rated movie languages
languagemean = languagemean.head()
#plot means of movie ratings according to language (only top 5)
ax2 = languagemean.plot.bar(rot=45, color='g')
plt.xlabel("Language")
plt.ylabel("Average IMDB Score")
plt.title("Languages of Netflix Original Movies With Highest IMDB Ratings")
plt.savefig("Figures/top5languagemeans.png")
```

```python
#only keep language & IMDB score
popular_language = netflix_origs_split[["Language", "IMDB Score"]]
#sort by language & count languages 
languages = popular_language.sort_values(by = "Language" )["Language"].unique()
#add extra column to assign numeric index to most popular languages
popular_language["Language_num"]=""
popular_language
```

```python
#assign numeric values to top 5 most popular languages
popular_language.loc[popular_language['Language'] == "English", 'Language_num'] = 1
popular_language.loc[popular_language['Language'] == "Spanish", 'Language_num'] = 2
popular_language.loc[popular_language['Language'] == "Hindi", 'Language_num'] = 3
popular_language.loc[popular_language['Language'] == "French", 'Language_num'] = 4
popular_language.loc[popular_language['Language'] == "Italian", 'Language_num'] = 5

#assign 0s to other languages & remove them 
popular_language.replace('', 0, inplace = True)
popular_language = popular_language.loc[popular_language["Language_num"] != 0]
popular_language
```

```python
#run ANOVA test
stats.f_oneway(popular_language['Language_num'], popular_language['IMDB Score'])
```

The null hypothesis is that there is no significant impact of the language on the means of IMDB Ratings. 
With a p-value of 0 it indicates the null hypothisis failed.

```python
#reused from language split but has errors  in the same code above. added extra delimeter to deal with occational cases
#has errors on hyphenated phrases ex. "coming-of-age"
lst = []
for i in range(len(netflix_originals)):
    genre = str(netflix_originals.loc[i]["Genre"])
    if ("/" in genre):
        split = genre.split("/")
        for j in range(len(split)):
            row = netflix_originals.loc[i].copy()
            row["Genre"] = split[j]
            lst.append(row)
    elif ("-" in genre):
        split = genre.split("-")
        for j in range(len(split)):
            row = netflix_originals.loc[i].copy()
            row["Genre"] = split[j]
            lst.append(row)
    else:
        row = netflix_originals.loc[i].copy()
        lst.append(row)
netflix_origs_genre = pd.DataFrame(lst)
# netflix_origs_genre = netflix_origs_genre.replace({'Genre': 'Horror thriller' , 'Horror anthology'})
netflix_origs_genre["Genre"] = netflix_origs_genre["Genre"].str.strip()
#restructre the genres into larger subgroups, these choices were based on personal opnion, changes could be made to further split these by spaces but could loose important data
netflix_origs_genre= netflix_origs_genre.replace(dict.fromkeys(['Science fiction','Science fiction adventure','Science fiction thriller','Science Fiction'], 'Sci Fi'))
netflix_origs_genre= netflix_origs_genre.replace(dict.fromkeys(['Horror thriller','Horror anthology','Horror comedy','Zombie','Teen comedy horror'], 'Horror'))
netflix_origs_genre= netflix_origs_genre.replace(dict.fromkeys(['Action comedy','Black comedy','Dark comedy','Comedy horror','Dance comedy','Comedy-drama',
                                                                'Musical comedy','Christmas comedy','Comedy mystery','camera prank comedy','Animated musical comedy'
                                                                ,'age comedy','Teen comedy'], 'Comedy'))
netflix_origs_genre= netflix_origs_genre.replace(dict.fromkeys(['Political thriller','thriller','Psychological thriller drama','Psychological thriller',
                                                                'Romantic thriller','Crime thriller','Spy thriller'], 'Thriller'))
netflix_origs_genre= netflix_origs_genre.replace(dict.fromkeys(['drama','Supernatural drama','Crime drama','War drama','Historical drama'], 'Drama'))
netflix_origs_genre= netflix_origs_genre.replace(dict.fromkeys(['Romantic comedy','Romance drama','Romantic teenage drama','romance','Romantic drama',
                                                                'Romantic teen drama'], 'Romance'))
netflix_origs_genre= netflix_origs_genre.replace(dict.fromkeys(['Christmas musical','Christian musical','Musicial'], 'Musical'))
netflix_origs_genre["Genre"].unique()
```
